```python
X,y = df.drop("Depression",axis=1),df['Depression']
```

```python
X = pd.get_dummies(X)
X.head(2)
```

```python
data_split = int(len(X)*0.8)
X = X.astype("float")
X_train,X_test,y_train,y_test = X[:data_split], X[data_split:], y[:data_split], y[data_split:]

X_train = X_train.reset_index(drop=True)
X_test = X_test.reset_index(drop=True)
y_train = y_train.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
X_train.head(2)
```

```python
import numpy as np

# Fungsi Sigmoid
def sigmoid(z):
    z = np.asarray(z)  # Memastikan z adalah array
    return 1 / (1 + np.exp(-z))

# Fungsi Binary Cross-Entropy
def binary_cross_entropy(X, y, w, b):
    m = len(X)  # Jumlah sampel
    prediction = np.dot(X, w) + b  # Prediksi (m,)
    prediction = sigmoid(prediction)  # Terapkan sigmoid
    
    # Menghitung error menggunakan binary cross-entropy
    epsilon = 1e-15
    prediction = np.clip(prediction,epsilon,1-epsilon)
    error = -(y * np.log(prediction) + (1 - y) * np.log(1 - prediction))
    error = np.sum(error) / m  # Rata-rata error per sampel
    return error

def hitung_gradient(X, y, w, b):
    m = len(X)
    dj_dw = 0
    dj_db = 0
    prediction = np.dot(X,w)+b
    prediction = sigmoid(prediction)
    error = prediction-y
    dj_dw = np.dot(error,X)/m
    dj_db = np.sum(error)/m

    return dj_dw,dj_db

def gradient_descent(X,y,w,b, epochs,alpha,step=1000):

    w_baru = w
    b_baru = b
    loss_hist = []
    for i in range(epochs):
        dj_dw, dj_db = hitung_gradient(X,y,w_baru,b_baru)
        w_baru = w_baru - alpha * dj_dw
        b_baru = b_baru - alpha * dj_db

        loss = binary_cross_entropy(X,y,w_baru,b_baru)
        loss_hist.append(loss)
        if i % step == 0:
            print(f"Epoch {i}, Loss = {loss}")
    
    print(f"Final Loss = {loss}")
    return loss_hist,w_baru,b_baru

def accuracy(y_pred, y_test):
    tp = np.sum(y_pred == y_test)
    return (tp/len(y_pred))
```

```python
w = np.ones(X_train.shape[1])
b = 0

loss_hist, w,b = gradient_descent(X_train,y_train,w,b,20000,0.1)
```

```python
prediksi = np.dot(X_test,w)+b
y_pred = sigmoid(prediksi)

y_pred = np.where(y_pred>0.5,1,0)
y_pred = pd.Series(y_pred)
```

```python
def confusion_matrix_manual(y_test, y_pred):
    TP = np.sum((y_test == 1) & (y_pred == 1)) 
    FP = np.sum((y_test == 0) & (y_pred == 1))  
    TN = np.sum((y_test == 0) & (y_pred == 0)) 
    FN = np.sum((y_test == 1) & (y_pred == 0))
    
    return TP, FP, TN, FN
```

```python
print(f"Test Accuracy : {accuracy(y_pred,y_test)}")

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(loss_hist)
plt.title('Learning Curve Logistic Regression')
plt.ylabel('Loss')
plt.xlabel('Epochs')

plt.subplot(1,2,2)
TP, FP, TN, FN = confusion_matrix_manual(y_test, y_pred)
conf_matrix = np.array([[TN, FP], [FN, TP]])
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Predicted 0', 'Predicted 1'], yticklabels=['True 0', 'True 1'])
plt.title('Confusion Matrix')
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```
